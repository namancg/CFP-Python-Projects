```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('data.csv')
yes = df[df.Attrition == 'Yes']
```

```python
education_data = {1: "Below College", 2: "College", 3: "Bachelor", 4: "Master", 5: "Doctor"}
education = yes.replace({"Education": education_data})
education['Education'].value_counts().plot(kind='pie', autopct='%.1f%%')
plt.show()
```

OBSERVATION : DIFFERENT COLORS SHOWS DIFFERENT TYPES OF EMPLOYEES COMING TO A COMPANY

```python
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.show()
```

OBSERVATION : THE COUNT OF PEOPLE WHO WANT TO WORK OVERTIME IS LESSER COMPARED TO PEOPLE WHO DONT WANT TO

```python
plt.figure(figsize=(8, 4))
plt.title('Environment Satisfaction')
plt.xlabel('Environment Satisfaction')
plt.ylabel('Number of Employees')
plt.hist(df.EnvironmentSatisfaction)
x = ["Low", "Medium", "High", "Very High"]
y = [1, 2, 3, 4]
plt.xticks(y, x)
plt.show()
```

OBSERVATION: WHEN THE ENVIRONMENT SATISFACTION IS HIGH, THE NUMBER OF EMPLOYEES ARE HIGH

```python
job_involvement = df["JobInvolvement"]
colors = ["red", "yellow", "green", "blue"]
plt.title("Job Involvement")
x = ["Low", "Medium", "High", "Very High"]
plt.pie(df.JobInvolvement.value_counts(), colors=colors, labels=x, startangle=100, autopct='%1.1f%%')
plt.show()
```

OBSERVATION: COLOR INDICATES ON WHAT LEVEL THE JOB INVOLVEMENT IS SHOWN BY THE EMPLOYEES

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
sns.distplot(df['PerformanceRating'], ax=ax[0])
sns.distplot(df['RelationshipSatisfaction'], ax=ax[1])
plt.show()
```

OBSERVATION: WHEN THE DENSITY OF THE EMPLOYEES IS MORE, PERFORMANCE INCREASES,WHEN THE DENSITY OF THE EMPLOYEES IS MORE, RELATIONSHIP SATISFACTION INCREASES

```python
f, axs = plt.subplots(1, 2, figsize=(14, 6))
seaborn.barplot(y="JobSatisfaction", x="PerformanceRating", hue="Attrition", data=df, ax=axs[0])
seaborn.barplot(y="WorkLifeBalance", x="PerformanceRating", hue="Attrition", data=df, ax=axs[1])
plt.show()
```

OBSERVATION: WHEN JOB SATISFACTION IS YES, PERFORMANCE ALSO INCREASES,WHEN JOB SATISFACTION IS NO, PERFORMANCE ALSO DECREASES

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.countplot(x="YearsAtCompany", hue="Attrition", data=df)
plt.show()
```

OBSERVATION: EMPLOYEES WHO HAVE WORKED FOR 1-4 YEARS ARE ATTRIED MORE THAN OTHERS
